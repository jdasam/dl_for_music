```python
import numpy as np
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
```

## 사인파 합성하기
- 원하는 주파수, 진폭으로 사인파 합성

- ![sin_wave](http://www.dsprelated.com/josimages_new/mdft/img386.png)


```python
amplitude = 0.5 # Digital Audio는 -1~1로 표현하니 소리 진폭의 최대값은 1
frequency = 440 # 주파수 Hz
duration = 3 # 초second 단위
sample_rate = 16000

total_number_of_samples = duration * sample_rate
time_frame = np.arange(total_number_of_samples)
time_frame_in_sec = time_frame/sample_rate

sine_wave = amplitude * np.sin(2*np.pi*frequency*time_frame_in_sec)

plt.plot(sine_wave[:100])
ipd.Audio(sine_wave, rate=sample_rate)
```

## 배음 합성하기
- 함수 정의하기

```python
# 사인파 합성 코드를 함수로 정리
def make_sine_wav(frequency, amplitude, duration, sample_rate=16000):
  total_number_of_samples = duration * sample_rate
  time_frame = np.arange(total_number_of_samples)
  time_frame_in_sec = time_frame/sample_rate

  return amplitude * np.sin(2*np.pi*frequency*time_frame_in_sec)

sine_wave_by_function = make_sine_wav(frequency=440, amplitude=0.5, duration=3)
```
