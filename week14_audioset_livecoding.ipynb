```python
import torchaudio
from collections import Counter
import torch
import copy
from pathlib import Path
import numpy as np
import random

class AudioSet:
  def __init__(self, wav_dir, csv_path, json_path, topk=20):
    self.wav_dir = Path(wav_dir)
    self.meta = pd.read_csv(csv_path, delimiter=', ', header=2)
    with open(json_path) as f:
      self.ontology = json.load(f)
    self.ontology = {x['id']:x for x in ontology}
    self.filter_meta_by_id()
    self.count_most_common_tags(topk)
    
  def get_wav_list(self):
    wav_list = list(self.wav_dir.glob("*.wav"))
    self.wav_list = wav_list
    self.check_wav_length()
    return wav_list
  
  def check_wav_length(self):
    invalid_indices = []
    for i, wav_fn in enumerate(self.wav_list):
      y, _ = torchaudio.load(wav_fn)
      if y.shape[1] < 16000 * 9:
        invalid_indices.append(i)
    # invalid_indices = [0, 4, 7, 9]
    for i in reversed(invalid_indices):
      del self.wav_list[i]
    print(f"Deleted {len(invalid_indices)} wav files")
    return self.wav_list
  
  def get_id_from_wav_list(self):
    self.get_wav_list()
    id_list = ['_'.join(x.stem.split('_')[:-1]) for x in self.wav_list ]
    '''
    id_list = []
    for x in self.wav_list:
      id = x.stem.split('_')[0]
      id_list.append(id)
    '''
    self.id_list = id_list
    return id_list
  
  def filter_meta_by_id(self):
    if not hasattr(self, 'id_list'):
      self.get_id_from_wav_list()
    mask = np.isin(self.meta['# YTID'].values, self.id_list)
    self.meta = self.meta[mask]
    return self.meta
  
  def count_most_common_tags(self, topk=20):
    total_tags = [self.ontology[y]['name'] for x in self.meta['positive_labels'].values for y in x.replace('"', '').split(',')]
    tag_counter = Counter(total_tags)
    most_common_tags = tag_counter.most_common(topk)
    self.selected_tags = [x[0] for x in most_common_tags]
    self.selected_tags = {x:i for i,x in enumerate(self.selected_tags)}
    return self.selected_tags
  
  def label_name_to_multi_hot_vector(self, labels):
    multihot = torch.zeros(len(self.selected_tags))
    
    for label in labels:
      multihot[self.selected_tags[label]] = 1
#     for label in labels:
#       idx = self.selected_tags.index(label)
#       multihot[idx] = 1
    return multihot

  def __len__(self):
    return len(self.meta)
      
  def __getitem__(self, idx):
    '''
    return audio_sample, label
    '''
    row = self.meta.iloc[idx]
    yt_id = row['# YTID']
    start_seconds = row['start_seconds']
    sample_path = self.wav_dir / f"{yt_id}_{start_seconds}.wav"
    
    y, sr = torchaudio.load(sample_path)
    y = y.mean(dim=0)
    start = random.randint(0,7999)
    y = y[start:start+int(16000*8.5)]
    
    labels = row['positive_labels'].replace('"', '').split(',')
    label_name = [self.ontology[x]['name'] for x in labels]
    
    label_name = [x for x in label_name if x in self.selected_tags]
    multihot = self.label_name_to_multi_hot_vector(label_name)
    
    return y, multihot

topk = 100

dataset = AudioSet('AudioSet/','balanced_train_segments.csv', 'ontology.json', topk)
trainset = copy.deepcopy(dataset)
validset = copy.deepcopy(dataset)

trainset.meta = trainset.meta[:int(len(dataset)*0.8)]
validset.meta = validset.meta[int(len(dataset)*0.8):]

from torch.utils.data import DataLoader

train_loader = DataLoader(trainset, shuffle=True, batch_size=128)
valid_loader = DataLoader(validset, shuffle=False, batch_size=128)
```

```python
import torch.nn as nn
class SpecModel(nn.Module):
  def __init__(self, sr, n_fft, hop_length, n_mels):
    super().__init__()
    self.mel_converter = torchaudio.transforms.MelSpectrogram(sample_rate=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
    self.db_converter = torchaudio.transforms.AmplitudeToDB()
  
  def forward(self, x):
    mel_spec = self.mel_converter(x)
    return self.db_converter(mel_spec)

class AudioModel(nn.Module):
  def __init__(self, sr, n_fft, hop_length, n_mels, hidden_size, num_output):
    super().__init__()
    self.sr = sr
    self.spec_converter = SpecModel(sr, n_fft, hop_length, n_mels)
    self.spec_norm = nn.BatchNorm1d(n_mels)
    self.conv_layer = nn.Sequential(
      nn.Conv1d(n_mels, out_channels=hidden_size//8, kernel_size=3),
      nn.BatchNorm1d(hidden_size//8),
      nn.MaxPool1d(2),
      nn.ReLU(),
      nn.Conv1d(hidden_size//8, out_channels=hidden_size//4, kernel_size=3),
      nn.BatchNorm1d(hidden_size//4),
      nn.MaxPool1d(2),
      nn.ReLU(),     
      nn.Conv1d(hidden_size//4, out_channels=hidden_size//2, kernel_size=3),
      nn.BatchNorm1d(hidden_size//2),
      nn.MaxPool1d(2),
      nn.ReLU(),
      nn.Conv1d(hidden_size//2, out_channels=hidden_size, kernel_size=3),
      nn.BatchNorm1d(hidden_size),
      nn.ReLU(),
    )
    self.final_layer = nn.Linear(hidden_size, num_output)

  def get_spec(self, x):
    '''
    Get result of self.spec_converter
    x (torch.Tensor): audio samples (num_batch_size X num_audio_samples)
    '''
    return self.spec_converter(x)
  
  def forward(self, x):
    spec = self.get_spec(x) # num_batch X num_mel_bins X num_time_bins
    spec = self.spec_norm(spec)
    out = self.conv_layer(spec)
    out = self.final_layer(out.permute(0,2,1))
    out = torch.max(out, dim=1)[0]
    out = torch.sigmoid(out)
    
#     out = torch.max(out, dim=-1)[0] # select [0] because torch.max outputs tuple of (value, index)
#     out = self.final_layer(out)
#     out = torch.sigmoid(out)
    return out
  
model = AudioModel(16000, 512, 256, 96, 64, topk)
```

```python
import tqdm
def train_model(model, train_loader, valid_loader, optimizer, num_epochs, loss_func, device='cuda'):
  loss_records =[] 
  valid_acc_records = []
  model = model.to(device)
  model.train() # Set model to train mode
  for epoch in tqdm.tqdm(range(num_epochs)):
    for batch in train_loader:
      optimizer.zero_grad() # Rest gradient of every parameters in optimizer (every parameters in the model)
      audio, label = batch
      audio = audio.to(device)
      label = label.to(device)
      pred = model(audio)
      loss = loss_func(pred, label.float())
      loss.backward() # Run backpropagation
      optimizer.step() # Update parameters
      loss_records.append(loss.item())
    valid_acc = validate_model(model, valid_loader, device)
    valid_acc_records.append(valid_acc.item())
  return {"loss": loss_records, "valid_acc": valid_acc_records}


def get_tpr_fpr(pred, target, threshold=0.5):
  thresh_pred = pred> threshold
  p = torch.sum(target == 1)
  tp = torch.sum((thresh_pred==1) * (target==1))
  n = torch.sum(target == 0)
  fp = torch.sum((thresh_pred==1) * (target==0))
  return tp/p, fp/n

def get_roc_auc(pred, label, num_grid=500):
  auc = 0
  prev_fpr = 0
  for thresh in reversed(torch.linspace(0,1,num_grid)):
    tpr, fpr = get_tpr_fpr(pred, label, threshold=thresh)
    auc += tpr * (fpr-prev_fpr)
    prev_fpr = fpr
  return auc

def validate_model(model, valid_loader, device, acc_func=get_roc_auc):
  valid_acc = 0
  model.eval()
  model.to(device)
  with torch.no_grad():
    for batch in valid_loader:
      audio, label = batch
      pred = model(audio.to(device))
      auc = acc_func(pred, label.to(device))
      valid_acc += auc * len(label)
  model.train()
  return valid_acc / len(valid_loader.dataset)
```

```python
def get_precision_recall(pred, target, threshold=0.5):
  thresh_pred = pred > threshold
  p = torch.sum(target == 1)
  pred_p = torch.sum(thresh_pred)
  tp = torch.sum((thresh_pred==1) * (target==1))
  if pred_p == 0:
    precision = 0
  else:
    precision = tp / pred_p
  if p == 0:
    recall = 0
  else:
    recall = tp / p
  return precision, recall

def get_pr_auc(pred, label, num_grid=500):
  auc = 0
  prev_recall = 0
  precisions = []
  recalls = []
  for thresh in reversed(torch.linspace(0,1,num_grid)):
    precision, recall = get_precision_recall(pred[...,:20], label[...,:20], threshold=thresh)
    auc += precision * (recall-prev_recall)
    prev_recall = recall
    precisions.append(precision)
    recalls.append(recall)
#   plt.plot(recalls, precisions)
  return auc

validate_model(model, valid_loader, device='cuda', acc_func=get_pr_auc)
```
